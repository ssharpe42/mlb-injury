```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import torch
from plotnine import *
from plotnine.themes import theme_538, theme_bw, theme_minimal, theme_xkcd

from injury.model.dataloader import PointProcessDataset
from injury.model.multihawkes import MultivariateHawkes
from injury.model.vocab import Vocab

pd.options.display.max_columns = 100
# %matplotlib inline

blog_path = (
    "/Users/Sam/Desktop/ssharpe42.github.io.nosync/assets/img/mlb-injury-point-process"
)
```

```python
def breaks(n):
    fun = lambda x: list(
        np.arange(int(np.ceil(x[0] / n)) * n, int(np.ceil(x[1] / n)) * n, n)
    )
    return fun
```

```python
def load_model(model):
    model_path = f"../logs/lightning_logs/{model}/checkpoints/"
    checkpoint = os.path.join(model_path, os.listdir(model_path)[0])
    # vocab = Vocab([])
    # vocab.load(f"../logs/lightning_logs/{model}/vocab.pkl")
    mh = MultivariateHawkes.load_from_checkpoint(checkpoint).eval()
    return mh.vocab, mh
```

```python
def extract_alpha(model, vocab):
    return pd.melt(
        pd.DataFrame(
            model.decay.log_alpha.weight.data.exp().squeeze().numpy()[1:-1, 1:-1],
            columns=vocab.vocab[1:],
            index=vocab.vocab[1:],
        )
        .rename_axis("prev_injury")
        .reset_index(),
        id_vars="prev_injury",
        var_name="next_injury",
        value_name="excitement",
    )
```

### Plot Excitement 

- all data dtd/il models
- all data combined models

```python
for player_type in ["batter", "pitcher", "all_players"]:
    for injury_type in ["dtd", "il", "both"]:

        model_file = player_type + ("_dtd" if injury_type != "both" else "") + "_all"
        vocab, mh = load_model(model_file)
        alpha_df = extract_alpha(mh, vocab)
        if injury_type == "il":
            title_event = "IL stints"
            plot_df = alpha_df[
                ~alpha_df.prev_injury.str.contains("dtd")
                & ~alpha_df.next_injury.str.contains("dtd")
            ]
        elif injury_type == "dtd":
            title_event = "day-to-day injuries"
            plot_df = alpha_df[
                alpha_df.prev_injury.str.contains("dtd")
                & alpha_df.next_injury.str.contains("dtd")
            ]
        else:
            title_event = "injuries"
            plot_df = alpha_df

        title_player_type = f"({player_type})" if player_type != "all_players" else ""
        p = (
            ggplot(
                plot_df,
                aes(y="prev_injury", x="next_injury", fill="excitement"),
            )
            + geom_tile()
            + scale_fill_continuous(name="Excitement\n(alpha)\n\n", cmap_name="Reds")
            + scale_x_discrete(name="\nNext Injury")
            + scale_y_discrete(name="Previous Injury\n")
            + theme_bw()
            + theme(
                figure_size=(7, 5),
                axis_text_x=element_text(rotation=90, size=10),
                axis_text_y=element_text(size=10),
                axis_title=element_text(size=12),
                legend_title=element_text(size=12),
            )
            + ggtitle(
                f"Excitement of past {title_event} on likelihood of "
                f"future {title_event} {title_player_type}"
            )
        )
        file = f"{player_type}-excitement-{injury_type}.png"
        p.save("../plots/" + file)
        p.save(os.path.join(blog_path, file))
        p
```

### Kernels

Visualize kernels of specific injuries and how intensity decays over time

```python
vocab, mh = load_model("pitcher_dtd_all")
```

```python
kernels = []
labels = []
days = []
n = 100
for injury1, injury2 in zip(
    ["elbow", "elbow (dtd)", "elbow (dtd)", "other arm", "foot"],
    ["elbow", "elbow", "elbow (dtd)", "elbow", "elbow"],
):
    idx1, idx2 = vocab.get_id(injury1), vocab.get_id(injury2)
    mu = mh.log_mu.weight.data.exp()[idx2]
    delta = mh.decay.log_delta.weight.data.exp()[idx1, idx2]
    alpha = mh.decay.log_alpha.weight.data.exp()[idx1, idx2]
    kernel = (mu + alpha * torch.exp(-delta * torch.arange(n))).numpy() * 162
    days.extend(list(range(n)))
    kernels.extend(list(kernel))
    labels.extend([injury1 + " --> " + injury2] * n)
```

```python
kernel_df = pd.DataFrame({"kernel": kernels, "injury": labels, "day": days})
p = (
    ggplot(kernel_df)
    + geom_line(aes(x="day", y="kernel", color="injury"), size=1.5)
    + scale_y_continuous(name="Intensity")
    + scale_x_continuous(name="Days")
    + scale_color_brewer(type="qual", palette=6)
    + theme_bw()
    + theme(
        figure_size=(7, 5),
        axis_text_x=element_text(size=10),
        axis_text_y=element_text(size=10),
        axis_title=element_text(size=12),
        legend_title=element_text(size=12),
    )
    + ggtitle(
        "Intensity kernels illustrate influence of past injuries on \nfuture elbow injuries"
    )
)
file = f"pitcher-elbow-kernel.png"
p.save("../plots/" + file)
p.save(os.path.join(blog_path, file))
p
```

```python
vocab, mh = load_model("batter_dtd_all")
```

```python
# Get sequences from injury events

player_id = 545361  # trout
player_data = data[data["player_id"] == player_id]
player_data["injury_ids"] = vocab(player_data["injury_location"])
seq = (
    player_data.groupby(["player_id", "name"], as_index=False)[["injury_ids", "t"]]
    .agg(list)
    .rename(columns={"injury_ids": "events", "t": "times"})
)
```

```python
# Process batch to feed into model

dl = PointProcessDataset(**seq[["events", "times"]], event_dim=len(vocab), has_eos=True)
tensors = dl.collate_fn([next(iter(dl))])
```

```python
# Get intensity each day

intensity = (
    torch.vstack(
        mh.predict_all(
            tensors["events"],
            tensors["times"],
            tensors["dt"],
            tensors["seq_lengths"],
            return_lambdas=True,
            max_dt=int(tensors["times"].max().item()),
            mc_samples=int(tensors["times"].max().item()),
        )
    )
    .detach()
    .numpy()
)
```

```python
# Scale intensity to season (180) and create data frame

intensity = pd.DataFrame(intensity[:, 1:-1] * 180, columns=vocab.vocab[1:])
intensity["Day"] = range(len(intensity))

intensity = pd.melt(
    intensity, id_vars="Day", var_name="Body Part", value_name="Intensity"
)
intensity["type"] = np.where(intensity["Body Part"].str.contains("dtd"), "DTD", "IL")
```

```python
# Actual events
event_times = pd.DataFrame(
    {
        "id": range(len(tensors["times"].squeeze().numpy())),
        "time": tensors["times"].squeeze().numpy().astype(int),
    }
)
event_times = event_times.merge(
    player_data[["date", "injury_location", "t"]].rename(columns={"t": "time"})
).query("injury_location!='[END]'")
event_times["type"] = np.where(
    event_times["injury_location"].str.contains("dtd"), "DTD", "IL"
)

# Total intensity for IL and by type
total_intensity = (
    intensity.query("type=='IL'")
    .groupby(["Day"], as_index=False)["Intensity"]
    .sum()
    .merge(event_times, left_on=["Day"], right_on=["time"], how="left")
)

total_intensity_by_type = (
    intensity.groupby(["Day", "type"], as_index=False)["Intensity"]
    .sum()
    .merge(event_times, left_on=["Day", "type"], right_on=["time", "type"], how="left")
)
```

```python
p = (
    ggplot(total_intensity[total_intensity["Day"].between(1200, 1600)])
    + geom_line(aes(x="Day", y="Intensity"), alpha=0.5)
    + geom_point(aes(x="time", y="Intensity", label="injury_location"))
    + geom_text(
        aes(x="time", y="Intensity", label="injury_location"),
        color="gray",
        adjust_text={
            "arrowprops": {"arrowstyle": "-", "color": "k", "lw": 0.5},
            "expand_text": (2, 2),
            "expand_points": (2, 2),
        },
    )
    + scale_x_continuous(name="Active Non-Injured Days", breaks=breaks(100))
    + scale_y_continuous(name="Intensity\n\n (Estimated Season IL Rate)")
    + theme_bw()
    + theme(
        figure_size=(7, 5),
        axis_text_x=element_text(size=10),
        axis_text_y=element_text(size=10),
        axis_title=element_text(size=12),
        legend_title=element_text(size=12),
        rect=element_blank(),
        axis_line_x=element_line(),
        axis_line_y=element_line(),
    )
    + ggtitle(title="Mike Trout IL Risk (2019 - 2022)")
)
file = f"trout-total-intensity.png"
p.save(os.path.join("../plots/", file))
p.save(os.path.join(blog_path, file))
p
```
